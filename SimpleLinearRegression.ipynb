Importing the Libraries


```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

Upload the DataSet and Read the csv


```python
data_set=pd.read_csv("Salary.csv")
```

Get the X and Y components from the csv and plot it

```python
X=data_set.iloc[:,:-1].values
y=data_set.iloc[:,-1].values
print(len(X),len(y))
```

```python
plt.scatter(X,y)
plt.plot(X,y)
```

Seperate the data_sets for training and testing

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=22)
```

Train the Model

```python
print(len(X_test)+len(X_train))
```

```python
from sklearn.linear_model import LinearRegression
regres=LinearRegression()
regres.fit(X_train,y_train)
```

Plotting the regression line

```python
plt.scatter(X_train,y_train)
plt.plot(X_train,regres.predict(X_train),color="red")
plt.xlabel("experience")
plt.ylabel("CTC")
plt.legend()
plt.title("SimpleLinearProgression")
```

Testing

```python
plt.scatter(X_test,y_test)
plt.plot(X_train,regres.predict(X_train),color="red")
plt.xlabel("experience")
plt.ylabel("CTC")
plt.legend()
plt.title("SimpleLinearProgression")
```
